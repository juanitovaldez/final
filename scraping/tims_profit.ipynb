```python
# Sample RO

RO_def = {'labor': {'rate': 149.00,
                    'hours': 5.3},
          'parts': {'price': 300.00,
                    'margin': 0.40,
                    'quantity': 2}}

payout_rates = {'service_labor': 0.10,
                'service_parts': 0.12,
                'tech': 0.28}
# Calculates Profit before carve outs
def get_profit(RO):
    labor_profit = RO['labor']['rate']*RO['labor']['hours']
    part_profit = (RO['parts']['price']*RO['parts']['margin']*RO['parts']['quantity'])/(1-RO['parts']['margin'])
    total_profit = labor_profit + part_profit
    profit_schedule = {'labor_profit': labor_profit,
                       'part_profit': part_profit,
                       'total_profit': total_profit}
    return(profit_schedule)

# Calculates Tech Payout and Service Writer payout and shops residual profit:
def payouts(profit_schedule, payout_schedule):
    service_pay = profit_schedule['labor_profit']*payout_schedule['service_labor'] + profit_schedule['part_profit']*payout_schedule['service_parts']
    tech_pay = profit_schedule['labor_profit']*payout_schedule['tech']
    shop_profit = profit_schedule['total_profit']-(service_pay+tech_pay)
    payout_schedule = {'service_writer': service_pay,
                       'tech': tech_pay,
                       'shop_profits': shop_profit}
    return payout_schedule
```

```python
profits = {'labor_profit': 351144,
            'part_profit': 299334,
            'total_profit': 349464.5}

payouts(profits, payout_rates)
        
```
